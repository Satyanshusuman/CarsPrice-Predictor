# Data cleaning:-

```python
counts=car_df["cars_name"].value_counts()
car_df=car_df[~car_df["cars_name"].isin(counts[counts==1].index)]
car_df
```

```python
car_df["cars_name"]=car_df["cars_name"].str.title()
car_df["cars_brand"]=car_df["cars_brand"].str.title()
car_df
```

```python
car_df["kms"]=car_df["kms"].str.replace(",","").astype(int)
    
```

```python
car_df
```

```python
car_df["gasoliene_type"]=car_df["gasoliene_type"].str.slice(0,6)
car_df["gasoliene_type"]=car_df["gasoliene_type"].str.split("G").str.get(0).replace("CN","CNG")
car_df
```

```python
car_df["price"]=car_df["price"].str.replace(",","").str[1:].astype(int)
car_df
```

```python
car_df.info()
```

```python
car_df=car_df.reset_index(drop=True)
car_df
```
