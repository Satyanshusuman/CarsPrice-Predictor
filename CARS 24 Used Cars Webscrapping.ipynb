```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
import pandas as pd
from bs4 import BeautifulSoup
```

```python
website = 'https://www.cars24.com/buy-used-cars-bangalore/'

driver = webdriver.Chrome(service="")
driver.get(website) 
time.sleep(5)

previous_height = driver.execute_script('return document.body.scrollHeight')
time.sleep(5)
while True:
    driver.execute_script('window.scrollTo(0,document.body.scrollHeight)')
    time.sleep(10)
    new_height = driver.execute_script('return document.body.scrollHeight')
    if new_height == previous_height:
        break
    previous_height = new_height
    time.sleep(10)
file = open('cars24_banglore.html', 'w', encoding='utf-8')
file.write(driver.page_source)
file.close()
driver.close()
```

```python
data = open('cars24_banglore.html','r')

soup=BeautifulSoup(data,"lxml")
```

```python
divs=soup.find_all("div","_2z-Yu")

cars_name = [div.find('div',{'class':'_1mlAq'}).find('h2',{'class':'_2lmIw'}).text[5:] for div in divs ]
len(cars_name)
```

```python
cars_brand = [div.find('div',{'class':'_1mlAq'}).find('h2',{'class':'_2lmIw'}).text[5:].split(' ')[0] for div in divs ] 

model_year = [div.find('div',{'class':'_1mlAq'}).text[:4] for div in divs ] 

model = [div.find('ul',{'class':'_1hOnS'}).text.strip("Automatic").strip("Manual") for div in divs ]

car_type = [div.find('ul',{'class':'_1hOnS'}).text for div in divs ]
car_type=["Manual" if "Manual" in i else "Automatic" for i in car_type ]

kms = [div.find('ul',{'class':'_13yb6'}).text.split(' ')[0] for div in divs ]

owner = [div.find('ul',{'class':'_13yb6'}).text.split(' ')[1][2:] for div in divs ]

gasoliene_type = [div.find('ul',{'class':'_13yb6'}).text.split(' ')[2][5:] for div in divs ]

emi_per_month = [div.find('div',{'class':'WNwWR'}).text.split("h")[0]+"h" for div in divs ]

price = [div.find('div',{'class':'_18ToE'}).find('span').text for div in divs ]

zero_downpayment = [div.find('div',{'class':'WNwWR'}).find('span').text for div in divs ]
```

```python
d={'cars_name':cars_name,
   'cars_brand':cars_brand,
   'model':model,
   'model_year':model_year,
   'car_type':car_type,
   'kms':kms,
   'owner':owner,
   'gasoliene_type':gasoliene_type,
   'emi_per_month':emi_per_month,
   'price':price,
   'zero_downpayment':zero_downpayment
}

df_banglore=pd.DataFrame(d)
```
